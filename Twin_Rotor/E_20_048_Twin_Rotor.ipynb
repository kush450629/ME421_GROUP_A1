# Task #1: Twin Rotor System Dynamic Model

## 1. Mathematical Derivation
The motion of the twin rotor system is modeled as a rigid body with constrained rotation. The system is described by the following state-space equations:

### 1.1 Kinematics
The orientation of the body frame relative to the inertial frame is given by the rotation matrix $R \in SO(3)$, governed by:
$$\dot{R}=\hat{\omega}R$$

The spatial angular velocity $\omega$ is related to the spatial angular momentum $\pi$ via the spatial inertia tensor $\mathbb{I}^R$:
$$\omega=(\mathbb{I}^{R})^{-1}\pi$$

### 1.2 Dynamics
The rate of change of spatial angular momentum is determined by the sum of external moments:
$$\dot{\pi}=\tau^{e}+\tau^{u}$$

* **Manipulatable (Control) Moments ($\tau^u$):** These are generated by the two rotors based on control inputs $u_1$ and $u_2$.
$$\tau^{u}=R\begin{bmatrix}1&0\\ 0&0\\ 0&1\end{bmatrix}\begin{bmatrix}\cos \alpha&-\cos \beta\\ \sin \alpha&-\sin \beta\end{bmatrix}\begin{bmatrix}u_{1}\\ u_{2}\end{bmatrix}$$

* **Unmanipulatable (Constraint) Moments ($\tau^e$):** This represents the constraint moment that prevents the rotation of the device about the $b_2$ axis.
$$\tau^{e}=R\begin{bmatrix}0\\ T_{2}\\ 0\end{bmatrix}$$

The term $T_2$ is defined by the body-fixed angular velocity $\Omega = R^T\omega$:
$$T_{2}=e_{2}^{T}(\Omega\times\mathbb{I}\Omega+\mathbb{I}\dot{\Omega})$$
