# Task #1: Derivation of the Twin Rotor Dynamic Model

To show that the twin rotor system can be modeled by the given state-space equations, we apply the principles of rigid body dynamics in the spatial (inertial) frame.

## 1. Kinematic Derivation
The orientation of the rotor arm is represented by a rotation matrix $R \in SO(3)$, which maps the body-fixed frame $\mathcal{B}$ to the inertial frame $\mathcal{C}$. The velocity of any point on the rigid body leads to the standard kinematic relation:
$$\dot{R} = \hat{\omega}R$$
where $\omega$ is the spatial angular velocity. This confirms the first part of the model.

## 2. Momentum and Dynamics
By definition, the spatial angular momentum $\pi$ is the product of the spatial inertia tensor and the spatial angular velocity:
$$\pi = \mathbb{I}^R \omega = (R \mathbb{I} R^T) \omega$$
Solving for $\omega$ gives the relation:
$$\omega = (\mathbb{I}^R)^{-1} \pi$$

### 2.1 Applying Euler's Equation
Newton-Euler equations for rotational motion state that the rate of change of angular momentum is equal to the sum of external moments acting on the center of mass:
$$\frac{d}{dt}(\pi) = \sum \tau$$

In this specific system, the total external moment is partitioned into two components:
1. **Control Moments ($\tau^u$):** Produced by the thrust and torque of the two rotors.
2. **Constraint Moments ($\tau^e$):** Produced by the physical bearings/pivots that restrict motion (specifically preventing rotation about the $b_2$ axis).

Substituting these into the momentum balance, we obtain the second required equation:
$$\dot{\pi} = \tau^e + \tau^u$$

## 3. Modeling Torque Components
### 3.1 Manipulatable Control ($\tau^u$)
The control part maps the scalar inputs $u_1, u_2$ through the rotor geometry ($\alpha, \beta$) into the spatial frame via $R$:
$$\tau^{u}=R\begin{bmatrix}1&0\\ 0&0\\ 0&1\end{bmatrix}\begin{bmatrix}\cos \alpha&-\cos \beta\\ \sin \alpha&-\sin \beta\end{bmatrix}\begin{bmatrix}u_{1}\\ u_{2}\end{bmatrix}$$

### 3.2 Unmanipulatable Constraint ($\tau^e$)
The constraint moment prevents rotation about $b_2$. Using the body-fixed angular velocity $\Omega = R^T \omega$, we show:
$$\tau^{e}=R\begin{bmatrix}0\\ T_{2}\\ 0\end{bmatrix}, \quad T_{2}=e_{2}^{T}(\Omega\times\mathbb{I}\Omega+\mathbb{I}\dot{\Omega})$$
