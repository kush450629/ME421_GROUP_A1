**Heat Transfer Cooling Problem**
-----------------
Given Data
-------------
Number of containers ($N$) = $40$ \
 Mass of milk per container ($m_{m\_unit}$) = $5\text{ kg}$ \
 Mass of container ($m_{c\_unit}$) = $2\text{ kg}$\
 Specific heat of milk ($C_m$) = $4000\text{ J/kg}\cdot\text{K}$\
 Specific heat of aluminum ($C_c$) = $921.1\text{ J/kg}\cdot\text{K}$\
 Initial Temperature ($T_i$) = $40^\circ\text{C}$\
 Final Temperature ($T_f$) = $15^\circ\text{C}$\
 Temperature difference ($\Delta T$) = $40 - 15 = 25\text{ K}$\
 Time allowed ($t$) = $2\text{ hours} = 7200\text{ seconds}$\


 Thermal Load Calculation
 ------------
Total Mass Calculations$$M_{milk} = N \times m_{m\_unit} = 40 \times 5 = 200\text{ kg}$$$$M_{container} = N \times m_{c\_unit} = 40 \times 2 = 80\text{ kg}$$Total Heat Removal ($Q$ Requirement)$$Q_{total} = (M_{milk} \cdot C_m \cdot \Delta T) + (M_{container} \cdot C_c \cdot \Delta T)$$$$Q_{total} = (200 \cdot 4000 \cdot 25) + (80 \cdot 921.1 \cdot 25)$$$$Q_{total} = 20,000,000\text{ J} + 1,842,200\text{ J}$$$$Q_{total} = 21,842,200\text{ Joules (or 21.84 MJ)}$$Required Cooling Capacity ($P_{req}$)$$P_{req} = \frac{Q_{total}}{t} = \frac{21,842,200}{7200}$$$$P_{req} \approx 3033.64\text{ Watts} \approx \mathbf{3.03\text{ kW}}$$
